Both datasets contain information about TV shows and movies from different streaming platforms. The data comes explicitly from the streaming services of Prime Video, Netflix, Hulu, and Disney+. This information is relavent because those are the four most popluar streaming services in the United States. 

The first dataset is the all_streaming.csv dataset which contains descriptive information about different movies and TV shows from the streaming platforms. This information includes title, type of entertainment, cast, etc. The second dataset is the all_genre.csv dataset. It contains information pertaining to the genre category of the movie/TV show. By combining the two datasets we will have a full picture of the movie/TV show and how it relates to other observations.
