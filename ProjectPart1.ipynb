## **Data Sources:**

We got our datasets from Kaggle.com. The link to the datasets can be found [here](https://www.kaggle.com/datasets/sc0v1n0/4-services-streaming-movies-and-tv). The first dataset is the all_streaming.csv dataset which contains descriptive information about different movies and TV shows from the streaming platforms. ## **Data Description:**

#### **1. all_streaming.csv**: 

This dataset has 22,998 observations and eleven variables. There are four variables with missing observations. The only variable of interest with missing data is duration_seconds. This variable is missing 482 observations which accounts for about 2% of the observations. Since only a small percentage of observations have missing data, we will keep them as missing. 

The variables of interest in the streaming dataste include:
* movie_or_serie: (categoical) classifies the observation as either Movie or TV Show
* title: (description) indicates the title of the observation
* release_year: (time) year the observation was released
* duration_seconds: (quantatative) total duration in minutes or seasons of the observation 
* channel_streaming: (categorical) categorizes the observation by which streaming service it is available


#### **2. all_genre.csv**: 

This dataset has 22,998 observations and one hundred and eleven variables. There are no columns with missing observations. So, none of the columns need modified due to null values.

The variables of interest in the genres dataste include:
* action: (binary) indicates whether the observation's genre is action
* comedy: (binary) indicates whether the observation's genre is comedy
* adventure: (binary) indicates whether the observation's genre is adventure
* animation: (binary) indicates whether the observation is an animated movie/TV show
* anime: (binary) indicates whether the observation's genre is anime
* drama: (binary) indicates whether the observation's genre is drama
* international: (binary) indicates whether the observation is an international movie/TV show
* kids: (binary) indicates whether the observation is a kid's movie/TV show
* horror: (binary) indicates whether the observation's genre is horror

```python
#read in the genres dataset
genres = pd.read_csv("Data/all_genre.csv")

#print("\nGenres:")
#print(genres.head())

#print out basic dataframe info
print(genres.shape)
print(genres.info()) 
print(genres.describe())
```

```python
#print out numer and percent of null values in each dataframe
def data_stats(listdf):
    for df in listdf:
        null = df.isnull().sum()
        print(null)
        print(null/len(df))
        print("\n")

streaming_null = [streaming]
genres_null = [genres]


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 120)
#data_stats(genres_null)
```

```python
#create subsets of the dataframes with only the variables of interest 

streaming_subset = streaming[["movie_or_serie", "title", "release_year", "duration_seconds", "channel_streaming"]]

genres_subset = genres[["action", "comedy", "adventure", "animation", "anime", "drama", "international", "kids", "horror"]]

print(f"Streaming columns {streaming.columns} and shape {streaming.shape}")
print(f"Genres columns {genres.columns} and shape {genres.shape}")
```

```python
#merge the datasets by column binding 

streaming_combines = pd.concat([streaming_subset, genres_subset], axis=1)

print(f"Combined dataframe columns {streaming_combines.columns} and shape {streaming_combines.shape}")
```
