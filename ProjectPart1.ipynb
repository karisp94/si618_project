## **Data Description:**

#### **1. all_streaming.csv**: 

This dataset has 22,998 observations and eleven variables. There are four variables with missing observations. The only variable of interest with missing data is duration_seconds. This variable is missing 482 observations which accounts for about 2% of the observations. Since only a small percentage of observations have missing data, we will keep them as missing. 

The variables of interest in the streaming dataste include:
* movie_or_serie: (categoical) classifies the observation as either Movie or TV Show
* title: (description) indicates the title of the observation
* release_year: (time) year the observation was released
* duration_seconds: (quantatative) total duration in minutes or seasons of the observation 
* channel_streaming: (categorical) categorizes the observation by which streaming service it is available


#### **2. all_genre.csv**: 

This dataset has 22,998 observations and one hundred and eleven variables. There are no columns with missing observations. So, none of the columns need modified due to null values.

The variables of interest in the genres dataste include:
* action: (binary) indicates whether the observation's genre is action
* comedy: (binary) indicates whether the observation's genre is comedy
* adventure: (binary) indicates whether the observation's genre is adventure
* animation: (binary) indicates whether the observation is an animated movie/TV show
* anime: (binary) indicates whether the observation's genre is anime
* drama: (binary) indicates whether the observation's genre is drama
* international: (binary) indicates whether the observation is an international movie/TV show
* kids: (binary) indicates whether the observation is a kid's movie/TV show
* horror: (binary) indicates whether the observation's genre is horror


## **Data Manipulation:**

```python
#Library Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read in datasets and merge rows and create an other_genre column for all genres not included in the ones we chose 
#also basic info() and describe() and missing values and percent of missing values and what we should do with missing values 

#read in the streaming dataset
streaming = pd.read_csv("Data/all_streaming.csv")

#print("\nStreaming:")
#print(streaming.head())

#print out basic dataframe info
print(streaming.shape)
print(streaming.info())
print(streaming.describe())

#check what the unique classifications are for the different streaming platofrms
print(streaming["channel_streaming"].unique())
#print(streaming[streaming["channel_streaming"] == "netflix-shows"]["movie_or_serie"].unique())


```

```python
#read in the genres dataset
genres = pd.read_csv("Data/all_genre.csv")

#print("\nGenres:")
#print(genres.head())

#print out basic dataframe info
print(genres.shape)
print(genres.info()) 
print(genres.describe())
```

```python
#print out numer and percent of null values in each dataframe
def data_stats(listdf):
    for df in listdf:
        null = df.isnull().sum()
        print(null)
        print(null/len(df))
        print("\n")

streaming_null = [streaming]
genres_null = [genres]


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 120)
#data_stats(genres_null)
```

```python
#create subsets of the dataframes with only the variables of interest 

streaming_subset = streaming[["movie_or_serie", "title", "release_year", "duration_seconds", "channel_streaming"]]

genres_subset = genres[["action", "comedy", "adventure", "animation", "anime", "drama", "international", "kids", "horror"]]

print(f"Streaming columns {streaming.columns} and shape {streaming.shape}")
print(f"Genres columns {genres.columns} and shape {genres.shape}")
```

```python
#merge the datasets by column binding 

streaming_combines = pd.concat([streaming_subset, genres_subset], axis=1)

print(f"Combined dataframe columns {streaming_combines.columns} and shape {streaming_combines.shape}")
```

```python
# check which streaming platforms are included
streaming_combines["channel_streaming"].unique()
```

```python
# rename streaming platforms to be shorter
streaming_combines["channel_streaming"] = streaming_combines["channel_streaming"].replace({
    "netflix-shows": "netflix",
    "hulu-movies-and-tv-shows": "hulu",
    "amazon-prime-movies-and-tv-shows": "amazon_prime",
    "disney-movies-and-tv-shows": "disney_plus"
})
```

```python
# make sure the names were changed correctly
streaming_combines["channel_streaming"].unique()
```

## **Data Visualization:**

```python
# release year distribution
plt.figure(figsize=(10,6))
plt.hist(streaming_combines["release_year"], bins=30, color="lightskyblue", edgecolor='black')
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.title("Distribution of Release Years")
plt.xticks(rotation=45)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# number of movies and TV shows by platform
counts = streaming_combines["channel_streaming"].value_counts()
plt.bar(counts.index, counts.values, color="lightskyblue", edgecolor='black')
plt.xlabel("Streaming Platform")
plt.ylabel("Count")
plt.title("Movies vs TV Shows by Platform")
plt.xticks(rotation=45)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# movies vs tv shows by platform
counts = streaming_combines.groupby(["channel_streaming", "movie_or_serie"]).size().unstack()
ax = counts.plot(kind="bar", figsize=(10,6), color=["lightskyblue", "palegreen"], edgecolor='black')
plt.xlabel("Streaming Platform")
plt.ylabel("Count")
plt.title("Movies vs TV Shows by Platform")
plt.xticks(rotation=45)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# identify genre columns
non_genre_cols = ["movie_or_serie", "title", "release_year",
                  "duration_seconds", "channel_streaming"]
genre_cols = [col for col in streaming_combines.columns if col not in non_genre_cols]

# sum genres by platform
genre_by_platform = streaming_combines.groupby("channel_streaming")[genre_cols].sum()

# get list of platforms
platforms = streaming_combines["channel_streaming"].unique()

# create a color palette (netflix = red, disney+ = blue, amazon prime = green, hulu = purple)
color_palette = {
    "netflix": "#E50914",
    "disney_plus": "#113CCF",
    "amazon_prime": "#00A8E1",
    "hulu": "#1CE783"
}

# create one plot per platform showing the top 5 genres per platform
for platform in platforms:
    
    # filter platform
    platform_df = streaming_combines[streaming_combines["channel_streaming"] == platform]
    
    # sum genres for that platform
    genre_totals = platform_df[genre_cols].sum().sort_values(ascending=False).head(5)

    # plot
    plt.figure()
    plt.bar(genre_totals.index, genre_totals.values, color=color_palette[platform], edgecolor="black")
    plt.title(f"Top 5 Genres on {platform}")
    plt.xlabel("Genre")
    plt.ylabel("Number of Titles")
    plt.xticks(rotation=45, ha="right")
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()
```
